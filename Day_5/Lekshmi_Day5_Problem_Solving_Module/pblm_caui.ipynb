## PART 1: Reading the Excel File and Writing it in a structured way to be read by program

```python
df=pd.read_excel('Mk_rain.xls',header=1,skiprows=range(9))           #Skipping the initial 9 lines
df
```

```python
stations=df.iloc[6:,0]                           # Read the stations
stations
```

```python
stations.iloc[0][0:9]=='DISTRICT:'             # Checking whether the string is equal to a given string
```

#### Find the rows where each district section starts

```python

dist=[]
indices=[]
for i in range(0,stations.shape[0]):
    
    if (isinstance(stations.iloc[i],str)):                     # Check whether it is a string
        if(stations.iloc[i][0:9]=='DISTRICT:'):                # Check whether the first 9 letters mention DISTRICT:
            dist.append(stations.iloc[i][10:])                 # Make a list of districts
            indices.append(i)                                  # Find the rows where district names are given
        
```

```python
# Make a copy removing initial empty rows
df1=df.iloc[6:,:]
df1
```

#### Subset for First District

```python
#df_dist1=df1.iloc[indices[0]:indices[1],:]
df_dist1=df1.iloc[indices[0]+1:indices[1]-1,:]                      # Subset the data for that district only
#print(df_dist1)
n_stations=df_dist1.shape[0]                                        # Find the number of stations in the district
n_stations
stations=df_dist1.loc[:,'MET.SUB/DISTRICT/STATION'].values          # Find the station names
stations

district=[dist[0]]*n_stations                                       # Create a list of districts repeating n_stations time

print(district[0]+'\t\t'+stations[0])                               # Print district and station of first station
```

```python
df_dist1
```

#### Picking the rainfall information of all stations of that district

```python
rfdata_dist=df_dist1.loc[:,1:31].reset_index()
rfdata_dist
```

#### Create a structured dataframe for the district

```python
df_out=pd.DataFrame()                                            # Create empty dataframe
df_out['MET_SUBDIVISION']=['JHARKHAND']*n_stations               # Create columns and give values to the columns
df_out['DISTRICT']=district                 # [dist[0]]*n_stations
df_out['STATION']=stations
df_out
```

```python
df_final=pd.concat([df_out,rfdata_dist],axis=1)   # Concat the station information and rainfall data together for that district
df_final
```
