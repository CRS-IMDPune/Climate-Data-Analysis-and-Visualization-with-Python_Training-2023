### 1. Read the text file using read() function
read() function reads the entire file
The 'r' means that the file should be opened in read mode (i.e. you will get an error if the file does not exist). 

Syntax of open() function is:
file_object = open('file_name', 'mode')



```python
f=open('ind99.prt','r')             # Opening the .prt file
f.read()                            # Reading the file completely
f.close()                           # Close the file
```

### 2. Read the text file using readline() function
readlines() function reads first line in the file

```python
f=open('ind105.prn')            # Reading the .prn file
f_lines=f.readline()            # Read the first line only
f_lines
#f.close()
```

```python
f=open('ind105.prn')            # Reading the .prn file
f_lines=f.readline(5)           # Read the first line of given size
f_lines
#f.close()
```

### 3. Read the text file using readlines() function
readlines() function reads all the lines in the file
Add an integer to the readlines() function to control the number of lines

```python
f=open('ind99.prt')                 # Open the file
f_lines=f.readlines()               # Read the entire lines in the data
f.close()
```

```python
# Print any particular line from the lines in the data
f_lines[100]
```

```python
# To print a range of lines from the data
f_lines[150:160]
```

```python
# prints the first 5 characters of the 200th line in the file
# Gives the line ID
f_lines[200][0:5]
```

```python
# prints the first 5 characters of the 200th line in the file
# Gives the Station name
f_lines[200][9:31]
```

```python
# Gives the Latitude of the station
f_lines[200][31:36]
```

```python
# Gives the Longitude of the station
f_lines[200][39:44]
```

```python
# Gives the Rainfall of first day of the station
f_lines[200][45:51]
```

```python
# Convert a string to float64 data type
np.float64(f_lines[200][45:51])
```

```python
# Prepare a index range for reading the rainfall data
index=np.arange(45,232,6)
index.shape
```

```python
index[1]
```

```python
# Read the rainfall data from a single line and save it in a list
data=[]
a=f_lines[200]                                  # 200th line of the data
for i in range(0,index.shape[0]-1):             # For every index
    tmp=np.float64(a[index[i]:index[i+1]])      # Read the data and convert to floating point numbers
    data.append(tmp)                            # Append it to a list
    del tmp
data
```

### 4. Writing a Text File with open() function

```python
fout=open("Text_File_Write_Example1.txt","w")           # Open a file for writing
fout.write("This is a sample text file")                # Write a string to the file
fout.close()                                            # Close the file
```

### 5. Appending a string to already existing file

```python
fout=open("Text_File_Write_Example1.txt","a")           # Open a file for writing
fout.write("This is a sample text file Line 2")         # Append a string to the file
fout.write("\tThis is a sample text file Line 3")       # Append a string to the file with tab delimit
fout.write("\nThis is a sample text file Line 4")       # Write a string to next line in the file
fout.close() 
```

### 6. Efficient Way of Text File Handling to Save Memory 

```python
station=[]                              # Define an empty list for station

with open('ind99.prt') as test_file:    # Opening the file with a name as test_file
    for line in test_file:              # Looping over each lines in the file
        #print (line) 
        station.append(line[9:31].strip())     # Pick the station name and strip the empty white spaces
#station                                        # Print the station
```

```python
station=[]                               # Define an empty list for station
with open('ind99.prt') as topo_file:     # Opening the file with a name as test_file
    i=0                                  # Setting a dummy index
    for line in topo_file:               # Looping over each lines in the file
        if (i < 10):                     # Only 10 lines are needed; Otherwise program will run until all lines are finished
            #print (line) 
            station.append(line[9:31].strip())   # Pick the station name and strip the empty white spaces
            i+=1                          # After reading each line add one to the dummy index
station
```
