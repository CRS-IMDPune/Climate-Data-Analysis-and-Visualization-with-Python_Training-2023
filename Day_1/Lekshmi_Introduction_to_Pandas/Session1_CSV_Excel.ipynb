### 1.4 Conditional selection
[Back to Sections](#Pandas)    

```python
# Choose the all data rows only for those days where daily rainfall> 500 mm
all_cond=df.loc[df['RAIN_DAILY(mm)']> 500.0]           
all_cond
```

```python
# Choose the all data rows only for those days where daily rainfall> 500 mm
multi_cond=df.loc[df['RAIN_DAILY(mm)']> 500.0]           
multi_cond
```

```python
# Choose the temperature and hourly rainfall data rows only for those days where daily rainfall> 500 mm
multi_cond1=df.loc[df['RAIN_DAILY(mm)']> 500.0,['TEMP(C)','RAIN_HOURLY(mm)']]           
multi_cond1
```

```python
# Choose the temperature and hourly rainfall data rows only for those days where daily rainfall> 500 mm and less than 600 mm
multi_cond2=df.loc[(df['RAIN_DAILY(mm)']> 500.0) & (df['RAIN_DAILY(mm)']< 600.0),['TEMP(C)','RAIN_HOURLY(mm)']]           
multi_cond2
```

```python
# Choose the temperature and hourly rainfall data rows only for those days where daily rainfall> 500 mm or less than 10 mm
multi_cond3=df.loc[(df['RAIN_DAILY(mm)']> 500.0) | (df['RAIN_DAILY(mm)']< 10.0),['TEMP(C)','RAIN_DAILY(mm)']]           
multi_cond3
```

### 1.8 Mapping
[Back to Sections](#Pandas)    

####  A map is a term, borrowed from mathematics, for a function that takes one set of values and "maps" them to another set of values. In data science we often have a need for creating new representations from existing data, or for transforming data from the format it is in now to the format that we want it to be in later.

#### map function
map() returns a new Series where all the values have been transformed by your function.

```python
# Make a subset of dataframe
df4=df.iloc[0:100,0:4]
df4

# Find the mean of temperature from the subset dataframe
mean_temp=df4['TEMP(C)'].mean()

# Mapping the temperature column with temperature-mean temperature value 
diff_temp=df4['TEMP(C)'].map(lambda p: p-mean_temp)
diff_temp
```

### 1.9 Grouping
[Back to Sections](#Pandas)    

```python
# Groups the data based on a column 
a=df4.groupby('TEMP(C)').groups
print(a)
```

```python
# Gives the number of times each temperature values have occurred in the dataframe
df4.groupby('TEMP(C)')['TEMP(C)'].count()
```

```python
# Gives the minimum and maximum daily rainfall values for each temperature group in the dataframe

print(df4.groupby('TEMP(C)')['RAIN_DAILY(mm)'].min())
print(df4.groupby('TEMP(C)')['RAIN_DAILY(mm)'].max())
```

agg() function lets you run a bunch of different functions on your DataFrame simultaneously

```python
#We can generate a simple statistical summary of the dataset
df4.groupby('TEMP(C)')['RAIN_DAILY(mm)'].agg([max,min,len])
```

### 1.11 Missing value Handling
[Back to Sections](#Pandas)   

In the above analysis we note very high values of mean and standard deviation in all variables.
This is due to the presence of missing values. In the file in some places the missing data is replaced
with -99999.0. This is not being considered as a missing value while calculating statistics in Pandas.

For that what we do is to replace such missing values with None or numpy.nan once we read the dataframe. The same idea can be used to replace missing values with some required value (like 0), if required, using the replace() function in pandas.

https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.replace.html

```python
df7=df.replace(-99999.0, np.nan)   # Replacing -99999.0 with nan values
df7.loc[100:140,:]
```

```python
#Before correction
df['TEMP(C)'].mean()

# After correction
df7['TEMP(C)'].mean()
```

Pandas provides some methods specific to missing data. 
To select NaN entries you can use pd.isnull() (or its companion pd.notnull()).

```python
# To select the NaN entries in dataframe
df7[pd.isnull(df7['TEMP(C)'])]
```

```python
# To select the non NaN entries in dataframe
df7[pd.notnull(df7['TEMP(C)'])]
```

Replacing missing values is a common operation. Pandas provides a really handy method for this problem: fillna(). 
fillna() provides a few different strategies for mitigating such data. 

```python
# We can replace each NaN with a string "Unknown" using the fillna() function
df8=df7['TEMP(C)'].fillna("Unknown") # Gives the temperature column replaced with 'Unknown'
print(df8.iloc[100:150])
```

### 1.13 Combining
[Back to Sections](#Pandas)   

Pandas has three core methods for doing this. In order of increasing complexity, these are concat(), join(), and merge(). 

The simplest combining method is concat(). Given a list of elements, this function will smush those elements together along an axis.

This is useful when we have data in different DataFrame or Series objects but having the same fields (columns). 

```python
# Creating two dataframes from our datasets
df_1=df7.iloc[0:20,:]
df_2=df7.iloc[30:50,:]

# Merging those dataframes
df_merge=pd.concat([df_1, df_2])
df_merge
```
